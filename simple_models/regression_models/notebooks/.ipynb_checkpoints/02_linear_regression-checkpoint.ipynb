> **Goal:** Train a linear regression model using gradient descent in PyTorch.  
> **Data:** Processed California Housing arrays saved by `01_EDA.ipynb`  
> **Story:** Load → Tensorize → Dataset → DataLoader → Model → Train → Evaluate → Save

## Step 0 — Imports
New imports compared to notebook 01 are marked with `# NEW`.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import torch                              # NEW — the core PyTorch library
import torch.nn as nn                     # NEW — nn = neural network module. Contains layers, losses, etc.
from torch.utils.data import Dataset      # NEW — base class we inherit from to build a custom dataset
from torch.utils.data import DataLoader   # NEW — wraps a Dataset and handles batching + shuffling automatically

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)            # NEW — sets PyTorch's random seed for reproducibility
# torch.manual_seed does for PyTorch what np.random.seed does for NumPy.
# Without this, weight initialisation is different every run → results change every time.
np.random.seed(RANDOM_SEED)

# --- Device setup ---
# PyTorch can run on CPU or GPU. This line automatically picks GPU if available,
# otherwise falls back to CPU. For this dataset, CPU is fast enough.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')
print('All imports successful.')
````

---
## Step 1 — Load Processed Data
We load the `.npy` files saved at the end of notebook 01.  
No re-splitting, no re-scaling — the work is already done.

````python
DATA_DIR = '../data/processed/'

# np.load() reads a .npy file back into a numpy array
# allow_pickle=False is a safety setting — we don't need pickle for plain float arrays
X_train = np.load(DATA_DIR + 'X_train.npy')
X_val   = np.load(DATA_DIR + 'X_val.npy')
X_test  = np.load(DATA_DIR + 'X_test.npy')
y_train = np.load(DATA_DIR + 'y_train.npy')
y_val   = np.load(DATA_DIR + 'y_val.npy')
y_test  = np.load(DATA_DIR + 'y_test.npy')

print('Loaded shapes:')
print(f'  X_train: {X_train.shape}  y_train: {y_train.shape}')
print(f'  X_val:   {X_val.shape}    y_val:   {y_val.shape}')
print(f'  X_test:  {X_test.shape}   y_test:  {y_test.shape}')
````

---
## Step 2 — Convert NumPy Arrays to PyTorch Tensors

### Why tensors?
NumPy arrays and PyTorch tensors are both multi-dimensional number arrays — they look and feel almost identical. The key difference is:
- NumPy arrays live on **CPU only** and have no gradient tracking
- PyTorch tensors can live on **CPU or GPU** and can track gradients for backpropagation

Everything that goes into a PyTorch model must be a tensor. So we convert now.

````python
# torch.tensor() creates a new tensor by copying data from a numpy array
# dtype=torch.float32 forces 32-bit float precision
# WHY float32? Neural networks use float32 by default. NumPy defaults to float64.
# If you mix float32 and float64 in PyTorch, you get a dtype mismatch error.
# Always explicitly set dtype=torch.float32 when converting from NumPy.

X_train_t = torch.tensor(X_train, dtype=torch.float32)
X_val_t   = torch.tensor(X_val,   dtype=torch.float32)
X_test_t  = torch.tensor(X_test,  dtype=torch.float32)

# .unsqueeze(1) reshapes the target vector
# y_train is shape (N,) — a 1D vector with N elements
# nn.Linear outputs shape (N, 1) — a 2D matrix with N rows and 1 column
# If shapes don't match, the loss function throws an error or silently broadcasts wrong
# .unsqueeze(1) inserts a new dimension at position 1: (N,) → (N, 1)
# Now both prediction and target are (N, 1) — shapes match.
y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
y_val_t   = torch.tensor(y_val,   dtype=torch.float32).unsqueeze(1)
y_test_t  = torch.tensor(y_test,  dtype=torch.float32).unsqueeze(1)

print('Tensor shapes:')
print(f'  X_train_t: {X_train_t.shape}  y_train_t: {y_train_t.shape}')
print(f'  X_val_t:   {X_val_t.shape}    y_val_t:   {y_val_t.shape}')
print()
print(f'  Data type: {X_train_t.dtype}')   # should print torch.float32
print(f'  Device:    {X_train_t.device}')  # should print cpu (or cuda if GPU available)
````

---
## Step 3 — Build a Custom Dataset Class

### Why do we need a Dataset class?
PyTorch's `DataLoader` (which handles batching) requires the data to be wrapped in a `Dataset` object. Think of the relationship like this:

- **Dataset** = a shelf of individual items. You can ask it: *"give me item number 42"*
- **DataLoader** = a conveyor belt that picks items from the shelf, groups them into batches, and optionally shuffles the order

To build a Dataset, you inherit from `torch.utils.data.Dataset` and implement exactly **two methods**:
- `__len__`  → tells Python how many items are in the dataset
- `__getitem__` → tells Python how to retrieve one item by index

````python
class HousingDataset(Dataset):
    # Dataset is the PyTorch base class. By writing 'class HousingDataset(Dataset)'
    # we are saying: HousingDataset IS a Dataset, inheriting all its behaviour.

    def __init__(self, X, y):
        # __init__ runs once when you create the object: HousingDataset(X_train_t, y_train_t)
        # We simply store the tensors as attributes of the object.
        # self.X and self.y are now accessible from every method in this class.
        self.X = X  # feature matrix tensor, shape (N, 8)
        self.y = y  # target vector tensor,  shape (N, 1)

    def __len__(self):
        # __len__ is called when Python needs to know the size of this dataset.
        # The DataLoader calls this to know how many total samples exist.
        # len(self.X) returns the number of rows — i.e., the number of samples N.
        return len(self.X)

    def __getitem__(self, idx):
        # __getitem__ is called when you request one specific sample by its index.
        # Example: dataset[42] returns the 42nd (X, y) pair.
        # The DataLoader calls this repeatedly to assemble a batch.
        # idx is the integer index of the sample requested.
        # We return a tuple (features, label) for that one sample.
        return self.X[idx], self.y[idx]


# --- Instantiate the three datasets ---
train_dataset = HousingDataset(X_train_t, y_train_t)
val_dataset   = HousingDataset(X_val_t,   y_val_t)
test_dataset  = HousingDataset(X_test_t,  y_test_t)

# --- Sanity checks ---
print(f'Train dataset length: {len(train_dataset)}')
# Fetch one sample manually to verify __getitem__ works
sample_X, sample_y = train_dataset[0]
print(f'Sample 0 — X shape: {sample_X.shape}, y shape: {sample_y.shape}')
# Expected: X shape torch.Size([8]), y shape torch.Size([1])
````

---
## Step 4 — Build DataLoaders

The DataLoader is the engine that feeds data to the model during training.  
It handles three things automatically: **batching**, **shuffling**, and **parallel loading**.

### Why batching?
You don't feed all 12,384 training samples to the model at once. Instead you feed small **batches** (e.g., 32 samples at a time). After each batch, you compute the loss and update the weights. This is called **Mini-batch Gradient Descent**.

**Analogy:** Imagine grading exams. Instead of reading all 12,384 exams and then writing one summary, you read 32 exams, write a partial summary, update your understanding, read the next 32, and so on. You converge to a good summary faster.

````python
BATCH_SIZE = 32
# Batch size is a hyperparameter — 32 is a standard default.
# Smaller batches = noisier gradient estimates but faster weight updates
# Larger batches = smoother gradient estimates but slower updates and more memory

train_loader = DataLoader(
    train_dataset,        # the Dataset object to load from
    batch_size=BATCH_SIZE,# how many samples per batch
    shuffle=True,         # shuffle=True: randomise the order every epoch
                          # WHY shuffle? So the model doesn't memorise the order of training samples.
                          # If data is ordered (e.g., all expensive houses first),
                          # the model would see a biased view each epoch.
    num_workers=0         # num_workers=0: load data in the main process (safest on Windows/Mac)
                          # num_workers>0 uses parallel workers to load data faster on Linux
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,        # shuffle=False for validation and test — order doesn't matter here
                          # we just want every sample evaluated once, deterministically
    num_workers=0
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=0
)

# --- Sanity check: inspect one batch ---
# iter() creates an iterator from the DataLoader
# next() pulls the first batch from that iterator
X_batch, y_batch = next(iter(train_loader))
print(f'One batch — X shape: {X_batch.shape}, y shape: {y_batch.shape}')
# Expected: X shape torch.Size([32, 8]), y shape torch.Size([32, 1])
# 32 = batch size, 8 = number of features, 1 = one target value per sample

# How many batches per epoch?
print(f'Batches per epoch (train): {len(train_loader)}')
# len(train_loader) = ceil(N_train / BATCH_SIZE) = ceil(12384 / 32) = 387 batches
````

---
## Step 5 — Define the Model

### What is nn.Linear?
`nn.Linear(in_features, out_features)` implements exactly the linear regression equation:
$$\hat{y} = Xw + b$$
where `w` is a weight vector of size `(in_features,)` and `b` is a scalar bias.  
PyTorch initialises `w` and `b` with small random values automatically.

### Why inherit from nn.Module?
Every model in PyTorch must inherit from `nn.Module`. It gives your model superpowers for free:
- Automatically tracks all parameters (weights and biases) → the optimizer can find them
- Enables `.to(device)` to move everything to GPU in one call
- Enables `model.train()` / `model.eval()` mode switching
- Enables `model.state_dict()` for saving/loading weights

````python
class LinearRegressionModel(nn.Module):
    # nn.Module is PyTorch's base class for all models
    # By inheriting from it, our model gets all the built-in PyTorch machinery

    def __init__(self, n_features):
        # Always call super().__init__() first — this initialises nn.Module's internals
        # Without this line, PyTorch won't track your parameters and nothing will work
        super(LinearRegressionModel, self).__init__()

        # nn.Linear(in, out) creates one fully connected (linear) layer
        # in_features  = n_features = 8 (one weight per input feature)
        # out_features = 1 (we predict a single number: the house price)
        # Internally it stores: weight matrix W of shape (1, 8) and bias b of shape (1,)
        self.linear = nn.Linear(n_features, 1)

    def forward(self, x):
        # forward() defines how data flows through the model
        # It is called automatically when you do: model(X_batch)
        # x is the input batch — shape (batch_size, n_features)
        # self.linear(x) computes: x @ W.T + b  for every sample in the batch
        # Output shape: (batch_size, 1)
        return self.linear(x)


# --- Instantiate the model ---
N_FEATURES = X_train.shape[1]  # = 8
model = LinearRegressionModel(n_features=N_FEATURES)

# .to(device) moves all model parameters to the target device (CPU or GPU)
# If device is 'cuda', this moves weights to GPU memory
# If device is 'cpu', this is a no-op (does nothing) — safe to always include
model = model.to(device)

# Print the model architecture and its parameter count
print(model)
print()

# Count total trainable parameters
# model.parameters() returns an iterator of all weight tensors
# p.numel() returns the number of elements in tensor p
# We only count parameters where requires_grad=True — i.e., learnable parameters
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f'Total trainable parameters: {total_params}')
# Expected: 9 = 8 weights (one per feature) + 1 bias

# Peek at the initial (random) weights
print()
print('Initial weights (w):', model.linear.weight.data)
print('Initial bias   (b):', model.linear.bias.data)
````

---
## Step 6 — Define Loss Function and Optimizer

### Loss function: MSELoss
Mean Squared Error is the standard loss for regression:
$$\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (\hat{y}_i - y_i)^2$$
This is the same as the lecture's cost function $E(w)$, just divided by $N$.

### Optimizer: SGD
The optimizer implements gradient descent — it updates the weights after each batch:
$$w \leftarrow w - \eta \cdot \nabla_w \text{Loss}$$
where $\eta$ is the **learning rate** — how big a step to take in the direction of the gradient.  
We use SGD (Stochastic Gradient Descent) here — the most fundamental optimizer.

````python
# nn.MSELoss() creates the Mean Squared Error loss function
# reduction='mean' (default) means: average the squared errors over the batch
# Other option: reduction='sum' — sums instead of averages (changes the scale of the loss)
criterion = nn.MSELoss()

# Learning rate — the most important hyperparameter in gradient descent
# Too large: weights overshoot the minimum and training diverges (loss goes to infinity)
# Too small: training converges extremely slowly
# 0.01 is a reasonable starting point for SGD on scaled data
LEARNING_RATE = 0.01

# torch.optim.SGD creates the SGD optimizer
# model.parameters() passes ALL learnable parameters (weights + bias) to the optimizer
# The optimizer holds a reference to these parameters and updates them in .step()
# lr=LEARNING_RATE sets the step size η
optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE)

print(f'Loss function: {criterion}')
print(f'Optimizer:     SGD with lr={LEARNING_RATE}')
````

---
## Step 7 — Training Loop

### The 5-Step Ritual
Every training iteration (one batch) follows this exact 5-step pattern — memorise it:

```
1. optimizer.zero_grad()    ← clear old gradients from previous batch
2. predictions = model(X)   ← forward pass: compute ŷ
3. loss = criterion(ŷ, y)   ← compute the loss
4. loss.backward()          ← backward pass: compute gradients ∂loss/∂w
5. optimizer.step()         ← update weights: w ← w - lr * gradient
```

### What is an Epoch?
One **epoch** = one complete pass through the entire training dataset.  
With 12,384 samples and batch size 32, one epoch = 387 batches.

````python
N_EPOCHS = 100
# 100 epochs means we loop through the full training data 100 times
# For linear regression on tabular data, this is usually enough to converge

# We record loss at every epoch for plotting the learning curve later
train_losses = []  # list to store average training loss per epoch
val_losses   = []  # list to store average validation loss per epoch

for epoch in range(N_EPOCHS):
    # =========================================================
    # TRAINING PHASE
    # =========================================================
    model.train()
    # model.train() puts the model in TRAINING mode.
    # For linear regression this doesn't change behaviour (no Dropout or BatchNorm here).
    # But it is essential to always include — it becomes critical for deeper models later.

    epoch_train_loss = 0.0
    # We accumulate the loss over all batches in this epoch
    # then divide by the number of batches to get the average

    for X_batch, y_batch in train_loader:
        # train_loader yields one (X_batch, y_batch) pair per iteration
        # X_batch shape: (32, 8)  — 32 samples, 8 features
        # y_batch shape: (32, 1)  — 32 target values

        # Move batch to the target device (CPU or GPU)
        # The model lives on 'device', so the data must be on the same device
        # Mixing CPU tensors with GPU tensors causes a runtime error
        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        # --- STEP 1: Clear gradients ---
        # PyTorch ACCUMULATES gradients by default — it adds new gradients to old ones.
        # This is useful in some advanced cases but harmful in standard training.
        # We must zero out the gradients before each new batch.
        # Without this, gradients from batch 1 + batch 2 + ... all pile up → wrong update.
        optimizer.zero_grad()

        # --- STEP 2: Forward pass ---
        # model(X_batch) calls model.forward(X_batch) internally
        # It computes: predictions = X_batch @ W.T + b
        # Output shape: (32, 1) — one predicted price per sample
        predictions = model(X_batch)

        # --- STEP 3: Compute loss ---
        # criterion computes MSE between predictions and ground truth labels
        # predictions shape: (32, 1)
        # y_batch shape:      (32, 1)
        # loss is a scalar tensor — a single number
        loss = criterion(predictions, y_batch)

        # --- STEP 4: Backward pass ---
        # .backward() computes the gradient of the loss w.r.t. every learnable parameter
        # After this call, model.linear.weight.grad and model.linear.bias.grad are populated
        # This is the chain rule applied automatically — PyTorch tracks every operation
        # in the forward pass and uses it to compute gradients in reverse
        loss.backward()

        # --- STEP 5: Update weights ---
        # .step() reads the gradients computed by .backward() and applies the update rule:
        # w ← w - lr * gradient
        # b ← b - lr * gradient
        # After this, the weights have moved one step in the direction that reduces loss
        optimizer.step()

        # Accumulate batch loss
        # .item() extracts the scalar Python float from the tensor
        # Without .item(), loss is still a tensor object — not a plain number
        epoch_train_loss += loss.item()

    # Average loss over all batches in this epoch
    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # =========================================================
    # VALIDATION PHASE
    # =========================================================
    model.eval()
    # model.eval() puts the model in EVALUATION mode.
    # Again, no behavioural difference for linear regression,
    # but critical habit for when you add Dropout or BatchNorm.

    epoch_val_loss = 0.0

    with torch.no_grad():
        # torch.no_grad() is a context manager that DISABLES gradient computation.
        # During validation we are not training — we don't need gradients.
        # Disabling them: (1) saves memory, (2) speeds up computation
        # Everything inside this block runs without building a computation graph
        for X_batch, y_batch in val_loader:
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)

            # Forward pass only — no backward, no optimizer step
            # We just want to see how good the current weights are on unseen data
            val_predictions = model(X_batch)
            val_loss = criterion(val_predictions, y_batch)
            epoch_val_loss += val_loss.item()

    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    # Print progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1:3d}/{N_EPOCHS}]  '
              f'Train MSE: {avg_train_loss:.4f}  '
              f'Val MSE:   {avg_val_loss:.4f}')

print()
print('Training complete.')
````

---
## Step 8 — Plot Learning Curves

The learning curve shows how train and val loss evolve over epochs.
It is the **most important diagnostic plot** in deep learning:
- Both curves decreasing → model is still learning
- Train loss decreasing but val loss increasing → overfitting
- Both curves flat from the start → learning rate too small or model is broken

````python
plt.figure(figsize=(10, 5))

# range(1, N_EPOCHS+1) creates [1, 2, 3, ..., 100] as the x-axis
# This makes the x-axis read '1' to '100' instead of '0' to '99'
epochs_range = range(1, N_EPOCHS + 1)

plt.plot(epochs_range, train_losses, label='Train MSE', color='steelblue',  linewidth=2)
plt.plot(epochs_range, val_losses,   label='Val MSE',   color='coral',      linewidth=2, linestyle='--')
# linestyle='--' draws the validation curve as a dashed line — easy to distinguish from train

plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Learning Curve — Linear Regression (California Housing)')
plt.legend()
plt.grid(True, alpha=0.3)  # alpha=0.3 makes the grid lines faint — readable but not distracting
plt.tight_layout()

os.makedirs('../results/plots', exist_ok=True)
plt.savefig('../results/plots/linear_regression_learning_curve.png', dpi=150, bbox_inches='tight')
plt.show()

print(f'Final Train MSE: {train_losses[-1]:.4f}')
print(f'Final Val   MSE: {val_losses[-1]:.4f}')
````

---
## Step 9 — Evaluate on Validation Set

We compute two metrics:
- **MSE** (Mean Squared Error) — the loss we trained on. Units are squared dollars.
- **RMSE** (Root MSE) — square root of MSE. Units are dollars — more interpretable.
  - $\text{RMSE} = \sqrt{\text{MSE}}$
  - Tells you: *"on average, my prediction is off by X dollars"*

````python
model.eval()
all_preds = []
all_true  = []

with torch.no_grad():
    for X_batch, y_batch in val_loader:
        X_batch = X_batch.to(device)
        preds = model(X_batch)

        # .cpu() moves the tensor from GPU back to CPU (needed if using CUDA)
        # .numpy() converts PyTorch tensor to NumPy array
        # .flatten() collapses shape (N, 1) to (N,) — a simple 1D array
        all_preds.append(preds.cpu().numpy().flatten())
        all_true.append(y_batch.numpy().flatten())

# np.concatenate joins a list of arrays into one long array
# e.g. [[1,2], [3,4], [5,6]] → [1, 2, 3, 4, 5, 6]
y_pred_val = np.concatenate(all_preds)
y_true_val = np.concatenate(all_true)

# Compute metrics manually so you understand what they calculate
mse  = np.mean((y_pred_val - y_true_val) ** 2)
rmse = np.sqrt(mse)
mae  = np.mean(np.abs(y_pred_val - y_true_val))
# MAE = Mean Absolute Error = average of |prediction - truth|
# Less sensitive to large outliers than MSE

print('=== Validation Set Metrics ===')
print(f'  MSE:  {mse:.4f}  (units: $100k squared)')
print(f'  RMSE: {rmse:.4f}  (units: $100k  → multiply by 100,000 for dollars)')
print(f'  MAE:  {mae:.4f}  (units: $100k)')
print()
print(f'  In plain English: predictions are off by ~${rmse*100000:,.0f} on average (RMSE)')
````

## Step 10 — Visualise Predictions vs Ground Truth

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Plot 1: Predicted vs True scatter plot ---
# A perfect model would put all points exactly on the diagonal line y=x
# Points above the line → model under-predicted
# Points below the line → model over-predicted
axes[0].scatter(y_true_val, y_pred_val, alpha=0.3, s=8, color='steelblue')

# Draw the perfect prediction line y=x
# min and max of the range give us start and end points of the diagonal
lims = [min(y_true_val.min(), y_pred_val.min()),
        max(y_true_val.max(), y_pred_val.max())]
axes[0].plot(lims, lims, 'r--', linewidth=1.5, label='Perfect prediction')
# 'r--' → red dashed line
axes[0].set_xlabel('True Value ($100k)')
axes[0].set_ylabel('Predicted Value ($100k)')
axes[0].set_title('Predicted vs True — Validation Set')
axes[0].legend()

# --- Plot 2: Residuals histogram ---
# Residual = prediction - truth (how wrong we were, and in which direction)
# A well-behaved model has residuals centred at 0 with roughly Gaussian shape
# Systematic bias (mean ≠ 0) means the model consistently over or under predicts
residuals = y_pred_val - y_true_val
axes[1].hist(residuals, bins=60, color='coral', edgecolor='white')
axes[1].axvline(0, color='black', linewidth=1.5, label='Zero residual')
axes[1].axvline(residuals.mean(), color='red', linestyle='--',
                label=f'Mean residual = {residuals.mean():.3f}')
axes[1].set_xlabel('Residual (Predicted − True)')
axes[1].set_ylabel('Count')
axes[1].set_title('Residuals Distribution')
axes[1].legend()

plt.suptitle('Linear Regression — Validation Diagnostics', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/linear_regression_diagnostics.png', dpi=150, bbox_inches='tight')
plt.show()
````

## Step 11 — Save Model Weights and Log Results

````python
os.makedirs('../results/models', exist_ok=True)
os.makedirs('../results/logs',   exist_ok=True)

# --- Save model weights ---
# model.state_dict() returns a dictionary of all parameter tensors:
#   {'linear.weight': tensor([[w1, w2, ..., w8]]), 'linear.bias': tensor([b])}
# torch.save() serialises this dictionary to a .pth file (PyTorch's native format)
# Convention: name the file to reflect the experiment — useful when you have many models
model_path = '../results/models/linear_regression_california.pth'
torch.save(model.state_dict(), model_path)
print(f'Model saved to: {model_path}')

# --- Log results to CSV ---
# We build one row of results as a dictionary, then append it to a CSV log file.
# This CSV will grow as you run more experiments (polynomial, ridge) — your ablation table.
log_path = '../results/logs/experiment_log.csv'

result_row = {
    'experiment':  'linear_regression',
    'dataset':     'california_housing',
    'model_type':  'OLS',
    'degree_M':    1,              # polynomial degree — 1 for plain linear
    'lambda':      None,           # regularisation strength — None for plain OLS
    'n_epochs':    N_EPOCHS,
    'batch_size':  BATCH_SIZE,
    'lr':          LEARNING_RATE,
    'train_mse':   round(train_losses[-1], 4),
    'val_mse':     round(val_losses[-1],   4),
    'val_rmse':    round(rmse,             4),
    'val_mae':     round(mae,              4),
    'test_mse':    None,           # we do NOT evaluate on test yet — only at the very end
}

# Convert the result row to a single-row DataFrame
result_df = pd.DataFrame([result_row])

# If log file already exists: append without writing a new header
# If log file does not exist yet: create it and write the header
if os.path.exists(log_path):
    result_df.to_csv(log_path, mode='a', header=False, index=False)
    # mode='a'     → append mode: adds rows to the bottom without overwriting
    # header=False → don't write column names again (they already exist in the file)
else:
    result_df.to_csv(log_path, mode='w', header=True, index=False)
    # mode='w'     → write mode: creates the file fresh
    # header=True  → write column names as the first row

print(f'Results logged to: {log_path}')
print()
print('Current experiment log:')
print(pd.read_csv(log_path).to_string())
````

## Step 12 — Inspect Learned Weights

````python
from sklearn.datasets import fetch_california_housing as _fetch
feature_names = _fetch().feature_names

# Extract learned weights from the model
# .detach() stops gradient tracking — we just want the values, not to compute gradients
# .numpy() converts to NumPy for easy display
weights = model.linear.weight.detach().numpy().flatten()  # shape (8,)
bias    = model.linear.bias.detach().numpy()[0]           # scalar

print('Learned model parameters after training:')
print(f'  Bias (b): {bias:.4f}')
print()
print('  Feature weights (sorted by absolute value):')

# Zip feature names with their weights and sort by absolute magnitude
weight_df = pd.DataFrame({'feature': feature_names, 'weight': weights})
weight_df['abs_weight'] = weight_df['weight'].abs()
weight_df = weight_df.sort_values('abs_weight', ascending=False).drop(columns='abs_weight')
print(weight_df.to_string(index=False))
print()
print('The feature with the largest absolute weight has the most influence on predictions.')
print('Does the top feature match what you expected from the correlation heatmap in EDA?')

# --- Bar chart of weights ---
plt.figure(figsize=(9, 4))
colours = ['steelblue' if w > 0 else 'coral' for w in weight_df['weight']]
# Positive weights → blue (feature pushes price up)
# Negative weights → red  (feature pushes price down)
plt.bar(weight_df['feature'], weight_df['weight'], color=colours, edgecolor='white')
plt.axhline(0, color='black', linewidth=0.8)
plt.xlabel('Feature')
plt.ylabel('Learned Weight')
plt.title('Linear Regression — Learned Feature Weights')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig('../results/plots/linear_regression_weights.png', dpi=150, bbox_inches='tight')
plt.show()
````
